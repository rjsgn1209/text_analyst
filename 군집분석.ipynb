```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from konlpy.tag import Hannanum, Komoran, Kkma, Twitter
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
import scipy.cluster.hierarchy as shc
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
han = Hannanum()
```

```python
data = pd.read_excel('data_files/clustering_data.xlsx')
data
```

```python
data.iloc[8,1]
```

```python
# 유사한 텍스트끼리 묶는다..
```

```python
kor = Komoran()

docs = []

for news in data['기사내용']:
    docs.append(kor.nouns(news))
```

```python
docs
```

```python
# 숫자로 전환..

# 1.
# countervectorizing

#         형태소1, 형태소2, 형태소3, 형태소4 ...
# 문서 1     0        2        1        0         =>(0,2,1,0, ...)
# 문서 2     1        1        0        0         =>(1,1,0,0, ...)         
# 문서 3     0        0        3        0         =>(0,0,3,0, ...)    
# ...

# 2.
# tfidf vectorizing
# 문서에서 형태소1의 등장수 : tf
# 전체 문서에서 형태소1이 등장한 문서수 : idf 
# tfidf = tf/idf

#         형태소1, 형태소2, 형태소3, 형태소4 ...
# 문서 1     0       0.3      0.01      0         =>(0, 0.3, 0.01, 0, ...)
# 문서 2   0.003    0.01       0        0         =>(0.003, 0.01, 0, 0, ...)         
# 문서 3     0        0       0.35        0         =>(0, 0, 0.35, 0, ...)    
# ...
```

```python
L = ['나라', '대한', '민국']
' '.join(L)
```

```python
# CountVectorizer에 값을 넣기위해 문장의 형식으로 변경
for i in range(len(docs)):
    docs[i] = ' '.join(docs[i])
```

```python
docs
```

```python
vec = CountVectorizer()
x = vec.fit_transform(docs)
x
```

```python
news_arr = x.toarray()
news_arr
```

```python
news_arr.shape
```

```python
news_arr[0] # aparse matrix => 학습이 잘 안됨.. 
```

```python
vec.get_feature_names()
```

```python
news_df = pd.DataFrame(news_arr, columns=vec.get_feature_names())
news_df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(news_df)
```

```python
kmeans.labels_
```

```python
cluster = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster.fit_predict(news_df)
```

```python
cluster = DBSCAN(eps=3, min_samples=5)
cluster.fit_predict(news_df)
```

```python
# 차원 축소
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
arr_pca = pca.fit_transform(news_df)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit_predict(arr_pca)
```

```python
vec = TfidfVectorizer(max_df=0.8, min_df=0.2)
x = vec.fit_transform(docs)
news_arr = x.toarray()
news_arr.shape
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(news_arr)
```

```python
kmeans.predict(news_arr)
```

```python
cluster = DBSCAN(eps=3, min_samples=5)
cluster.fit_predict(news_arr)
```
