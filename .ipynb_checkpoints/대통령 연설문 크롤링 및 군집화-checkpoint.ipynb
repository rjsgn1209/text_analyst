```python
text_li = []
President_li = ['김영삼', '김대중', '노무현', '이명박', '박근혜']
for i in range(5):
    driver.execute_script("window.scrollTo(0, 0)") 
    driver.find_element_by_css_selector(f'#prstNm{i+7}').click()
    driver.find_element_by_css_selector('#frm > fieldset.board-search2 > div.search-layer > div.buttons > button.button.color1.size1').click()
    time.sleep(2)
    for j in range(5):
        elem = driver.find_element_by_css_selector(f'#M_More > tr:nth-child({j+1}) > td.subject > a')
        elem.click()
        time.sleep(2)
        text = driver.page_source
        soup = BeautifulSoup(text, 'html.parser')
        text_li.append([President_li[i], soup.select('.content')[0].text.replace('\n', '').replace('\xa0', '')])
        print(President_li[i] + str(j+1), 'crawling')
        time.sleep(1)
        driver.back()
        time.sleep(2)
```

```python
df = pd.DataFrame(text_li, columns=['대통령', '연설문'])
```

```python
df
```

```python
kor = Komoran()
docs = []

for text in df['연설문']:
    docs.append(kor.nouns(text))
```

```python
docs
```

```python
count = 0
prst_text = []
x = []
for i in range(len(docs)):
    x += docs[i]
    if i%5 == 4:
        prst_text.append(x)
        count += 1
    
```

```python
word_count = pd.Series(prst_text[1]).value_counts()
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure(figsize=(20,5))
word_count[:20].plot(kind='bar')
plt.show()
```

```python
for i in range(len(docs)):
    docs[i] = ' '.join(docs[i])
```

```python
len(docs)
```

```python
vec = TfidfVectorizer()
```

```python
x = vec.fit_transform(docs)
text_arr = x.toarray()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit_predict(text_arr)
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
cluster.fit_predict(text_arr)
```

```python
cluster = DBSCAN(eps=3, min_samples=5)
cluster.fit_predict(text_arr)
```

```python
vec = CountVectorizer(max_df=0.7, min_df=0.2)
```

```python
x = vec.fit_transform(docs)
text_arr = x.toarray()
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit_predict(text_arr)
```

```python
cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
cluster.fit_predict(text_arr)
```

```python
cluster = DBSCAN(eps=3, min_samples=5)
cluster.fit_predict(text_arr)
```
