```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
train_df = pd.read_csv('data_files/nsmc-master/ratings_train.txt', sep='\t')
test_df = pd.read_csv('data_files/nsmc-master/ratings_test.txt', sep='\t')
```

```python
train_df.dropna(inplace=True)
test_df.dropna(inplace=True)
```

```python
from konlpy.tag import Komoran
```

```python
kor = Komoran()
```

```python
def preprocessing(doc):
    text = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ ]', ' ', doc)
    return text
```

```python
train_df['document1'] = train_df.document.apply(preprocessing)
```

```python
test_df['document1'] = test_df.document.apply(preprocessing)
```

```python
train_df
```

```python
# stopwords-ko.txt
```

```python
stw = pd.read_csv('data_files/stopwords-ko.txt', header=None)
stopwords = list(stw.iloc[:, 0].values)
stopwords
```

```python
def remove_stopwords(doc):
    text = [w for w in doc.split() if w not in stopwords]
    return ' '.join(text)
```

```python
train_df['document2'] = train_df.document1.apply(remove_stopwords)
```

```python
test_df['document2'] = test_df.document1.apply(remove_stopwords)
```

```python
train_df
```

```python
train_df = train_df[train_df.document2.apply(lambda x: x != '')]
```

```python
test_df = test_df[test_df.document2.apply(lambda x: x != '')]
```

```python
train_docs = []

for d in train_df['document2']:
    #print(d)
    tokens = kor.nouns(d)
    train_docs.append(tokens)
```

```python
test_docs = []

for d in test_df['document2']:
    #print(d)
    tokens = kor.nouns(d)
    test_docs.append(tokens)
```

```python
for i in range(len(train_docs)):
    train_docs[i] = ' '.join(train_docs[i])
```

```python
for i in range(len(test_docs)):
    test_docs[i] = ' '.join(test_docs[i])
```

```python
train_df['train_docs'] = train_docs
```

```python
test_df['test_docs'] = test_docs
```

```python
train_df = train_df[train_df['train_docs'] != '']
test_df = test_df[test_df['test_docs'] != '']
```

```python
train_docs = train_df.train_docs
test_docs = test_df.test_docs
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vec = CountVectorizer(min_df=3, max_df=0.9, max_features=2000)
x_train = vec.fit_transform(train_docs)
x_train_vec = x_train.toarray()
```

```python
vec = CountVectorizer(min_df=3, max_df=9, max_features=2000)
x_test = vec.fit_transform(test_docs)
x_test_vec = x_test.toarray()
```

```python
x_train_vec.shape
```

```python
x_test_vec.shape
```

```python
y_train = train_df['label']
y_test = test_df['label']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
log.fit(x_train_vec, y_train)
```

```python
preds = log.predict(x_test_vec)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, preds))
```
